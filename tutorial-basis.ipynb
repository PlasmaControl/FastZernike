```python
from zernipy.basis import ZernikePolynomial
from zernipy.plotting import plot_basis
```

```python
basis = ZernikePolynomial(L=5, M=5, spectral_indexing="ansi", sym=False)
plot_basis(basis);
```

```python
basis = ZernikePolynomial(L=5, M=5, spectral_indexing="ansi", sym="cos")
plot_basis(basis);
```

```python
basis = ZernikePolynomial(L=5, M=5, spectral_indexing="ansi", sym="sin")
plot_basis(basis);
```

```python
basis = ZernikePolynomial(L=5, M=5, spectral_indexing="fringe", sym=False)
plot_basis(basis);
```

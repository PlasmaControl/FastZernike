```python
import numpy as np
from zernipy.basis import ZernikePolynomial
from zernipy.plotting import plot_basis, plot_mode, plot_modes
import matplotlib.pyplot as plt
```

```python
modes = np.array([[6,2]])
modes = np.array([[6,2], [1,1], [9,3]])
plot_modes(modes, rho=100, theta=100, figsize=(8,8));
plt.savefig('zernike_modes.png', dpi=1200)
```

```python
plot_mode(mode=[6,2], rho=100, theta=100, figsize=(8,8));
plt.savefig('zernike_mode.png', dpi=1200)
```

```python
basis = ZernikePolynomial(L=5, M=5, spectral_indexing="ansi", sym=False)
plot_basis(basis);
```

```python
basis = ZernikePolynomial(L=5, M=5, spectral_indexing="ansi", sym="cos")
plot_basis(basis);
```

```python
basis = ZernikePolynomial(L=5, M=5, spectral_indexing="ansi", sym="sin")
plot_basis(basis);
```

```python
basis = ZernikePolynomial(L=5, M=5, spectral_indexing="fringe", sym=False)
plot_basis(basis);
```
